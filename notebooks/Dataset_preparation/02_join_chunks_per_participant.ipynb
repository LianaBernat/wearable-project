```python
#Função para empacotar
import pandas as pd
from pathlib import Path

METADATA_PATH = "../../data/data_raw/capture24/metadata.csv"
ALL_PARTICIPANT_PATH ="../../data/data_processed/participants/"
metadata = pd.read_csv(METADATA_PATH)

def join_chunks(metadata, all_participant_path):
    pid = metadata['pid']
    for id in pid:
        participant_path = f'{ALL_PARTICIPANT_PATH}{id}'
        participant_path = Path(participant_path)
        chunk_files = sorted(participant_path.glob("chunk_*.parquet"))

        dfs = []
        for f in chunk_files:
            df = pd.read_parquet(f)
            dfs.append(df)

        full_df=pd.concat(dfs, ignore_index=True)
        output_path= participant_path/f"{id}_full.parquet"
        full_df.to_parquet(output_path, index=False)
        print(f"[OK] {id} chunks successfully joined → {output_path}")
```

```python
#teste para dois participantes
metadata_test =metadata.iloc[0:2,:]
metadata_test['pid']
```

```python
#teste para dois participantes
join_chunks(metadata_test, ALL_PARTICIPANT_PATH)
```
