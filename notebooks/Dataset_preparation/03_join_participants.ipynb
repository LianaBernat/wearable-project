```python
#Função para empacotar
import pandas as pd
from pathlib import Path

METADATA_PATH = "../data/data_raw/capture24/capture24/metadata.csv"
ALL_PARTICIPANT_PATH ="../data/data_processed/participants/"
metadata = pd.read_csv(METADATA_PATH)

def join_participants(metadata, all_participant_path):
    pid = metadata['pid']
    dfs = []

    for id in pid:
        participant_path = f'{ALL_PARTICIPANT_PATH}{id}'
        participant_file = Path(participant_path)/f"{id}_full.parquet"

        df = pd.read_parquet(participant_file)
        dfs.append(df)
        print(f"[OK] {id} successfully joined")

    full_df=pd.concat(dfs, ignore_index=True)
    output_path= Path(all_participant_path)/f"Participants_all.parquet"
    full_df.to_parquet(output_path, index=False)
    return output_path
```
