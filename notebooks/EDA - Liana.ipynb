```python
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
pd.set_option('display.max_row', None)
```

```python
ALL_PARTICIPANT_PATH ="../data/data_processed/participants/"
```

```python
df_all = pd.read_parquet((Path(ALL_PARTICIPANT_PATH)/'Participants_all.parquet'))
```

```python
#observaÃ§Ãµes por participante
n_participants = (df_all['pid']).value_counts()

max_n_part = max(n_participants)
min_n_part = min(n_participants)

ax = sns.barplot(n_participants)

ax.axhline(max_n_part, color='red', linestyle='--', linewidth=2, label=f'Max: {max_n_part:,.2f}')
ax.axhline(min_n_part, color='orange',linestyle='--', linewidth=2, label=f'Min: {min_n_part:,.2f}')

ax.set_title('Number of observations per participant')

ax.legend()

ax.set_xlabel('Participant')
ax.set_ylabel('Observations')
plt.xticks(rotation=90, ha='right')

for index, label in enumerate(ax.get_xticklabels()):
    if index % 10 != 0:
        label.set_visible(False)

plt.show()
```

```python
# Gender and age distribution
gender_counts = df_all['sex'].value_counts(dropna=False)
age_counts = df_all['age_group'].value_counts().sort_index()


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

sns.barplot(x=gender_counts.index, y=gender_counts.values, hue = gender_counts, palette='Set2',ax=ax1)
ax1.set_title('Gender Distribution')
ax1.set_xlabel('Gender: 0 - man, 1-woman')
ax1.set_ylabel('Count')


sns.barplot(x=age_counts.index,y=age_counts.values,ax=ax2,color='skyblue')
ax2.set_title('Age group distribution')
ax2.set_xlabel('Age Group')
ax2.set_ylabel('Count')

label =['0: 18â€“29 anos','1: 30â€“37 anos','2: 38â€“52 anos','3: 53+ anos']

ax2.legend(label, title="Faixas etÃ¡rias", loc="lower right")

plt.tight_layout()
plt.show();
```
